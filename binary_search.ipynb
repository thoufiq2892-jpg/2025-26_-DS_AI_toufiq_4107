```python
import pandas as pd
data={
"Name": ['John', 'Alice', 'Jay', 'Mark'], "Age": [20,21,22,19],
"Marks": [70,85,75,90]}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Filtering data
print("Alice's data:")
print(df [df["Name"]=='Alice'])
print("Students with marks less than 85:")
print(df [df["Marks"]<85])
print("Students with marks greater than 75 and age less than 20:")
print(df[(df["Age"]>20)&(df["Marks"]>75)])
```

```python
#selection sort
n = int(input("Enter number of elements: "))
arr = []

# Taking input
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("Original array:", arr)

# Selection Sort logic
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    # Swap the found minimum element with the current element of the unsorted part
    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted array:", arr)
```

```python
#bubble sort
n = int(input("Enter number of elements: "))
arr = []

# Taking input
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("Original array:", arr)

# Bubble sort
for i in range(n - 1):
    for j in range(n - 1 - i):
        if arr[j] > arr[j + 1]:
            # Swap
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print("Sorted array:", arr)
```

```python
# Linear search
n = int(input("Enter number of elements: "))
arr = []

# Taking input
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

key = int(input("Enter element to search: "))

# Linear search logic
found = False
for i in range(n):
    if arr[i] == key:
        print(f"Element {key} found at position {i}")
        found = True
        break

if not found:
    print(f"Element {key} not found in the array")
```

```python
# Binary Search
n = int(input("Enter number of elements: "))
arr = []

# Input elements
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

# Sort the list first (binary search requires a sorted list)
arr.sort()
print("Sorted list:", arr)

key = int(input("Enter element to search: "))

# Binary search logic
low = 0
high = n - 1
found = False

while low <= high:
    mid = (low + high) // 2
    if arr[mid] == key:
        print(f"Element {key} found at position {mid}")
        found = True
        break
    elif arr[mid] < key:
        low = mid + 1
    else:
        high = mid - 1

if not found:
    print(f"Element {key} not found in the list")
```
