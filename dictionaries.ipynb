```python
# Create a dictionary
student = {
    "name": "Alice",
    "age": 20,
    "course": "CS"
}

# Access values
print(student["name"])

# Add a new key-value pair
student["grade"] = "A"

# Update a value
student["age"] = 21

# Remove a key-value pair
del student["course"]

# Length of dictionary
print(len(student))

# Membership test (check if a key exists)
print("name" in student )
```
