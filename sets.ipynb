```python
# Create sets
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Add an element
A.add(7)

# Remove an element
A.remove(2)

# Length of set
print(len(A))

# Union
union_set = A | B

# Intersection
intersection_set = A & B

# Difference
difference_set = A - B

# Membership test
print(3 in A)

# Loop through set
for item in A:
    print(item)

# Display all sets
print("A:", A)
print("Union:", union_set)
print("Intersection:", intersection_set)
print("Difference:", difference_set)
```
