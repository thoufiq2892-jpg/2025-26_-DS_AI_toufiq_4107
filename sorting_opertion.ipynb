```python
import pandas as pd
data={
"Name": ['John', 'Alice', 'Jay', 'Mark'], "Age": [20,21,22,19],
"Marks": [70,85,75,90]}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Filtering data
print("Alice's data:")
print(df [df["Name"]=='Alice'])
print("Students with marks less than 85:")
print(df [df["Marks"]<85])
print("Students with marks greater than 75 and age less than 20:")
print(df[(df["Age"]>20)&(df["Marks"]>75)])
```

```python
#selection sort
n = int(input("Enter number of elements: "))
arr = []

# Taking input
for i in range(n):
    arr.append(int(input(f"Enter element {i+1}: ")))

print("Original array:", arr)

# Selection Sort logic
for i in range(n - 1):
    min_index = i
    for j in range(i + 1, n):
        if arr[j] < arr[min_index]:
            min_index = j

    # Swap the found minimum element with the current element of the unsorted part
    arr[i], arr[min_index] = arr[min_index], arr[i]

print("Sorted array:", arr)
```
