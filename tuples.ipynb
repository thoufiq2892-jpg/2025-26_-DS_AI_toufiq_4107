### What is a Tuple?
A tuple is an ordered, immutable collection of items in Python. This means once a tuple is created, its elements cannot be changed. Tuples are defined by enclosing the elements in parentheses `()`.

```python
# Create a tuple
my_tuple = (10, 20, 30)

# Access elements
print(my_tuple[0])

# Length of tuple
print(len(my_tuple))

# Concatenation
new_tuple = my_tuple + (40, 50)

# Repetition
repeat_tuple = my_tuple * 2

# Membership test
print(20 in my_tuple)

# Loop through tuple
for item in my_tuple:
    print(item)

# Display new tuples
print(new_tuple)
print(repeat_tuple)
```
