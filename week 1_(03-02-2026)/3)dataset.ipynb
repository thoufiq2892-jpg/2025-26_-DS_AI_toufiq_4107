```python
#3. Write the code to download, save, and transform the dataset from a given URL.
import pandas as pd

# URL of dataset (example CSV)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"

# Download data
df = pd.read_csv(url)

# Save dataset locally
df.to_csv("iris_dataset.csv", index=False)

print("Dataset downloaded and saved successfully!")

# ---- Transformations ----
# Display first 5 rows
print(df.head())

# Add a new column (example transformation)
df["sepal_area"] = df["sepal_length"] * df["sepal_width"]

# Save transformed data
df.to_csv("iris_transformed.csv", index=False)

print("Dataset transformed and saved!")
```
