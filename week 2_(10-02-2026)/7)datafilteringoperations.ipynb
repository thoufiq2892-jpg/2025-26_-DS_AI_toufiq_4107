```python
#FILTERING DATA FROM DICTIONARIES
import pandas as pd
data={
    "Name":['John','Alice','Jay','Mark'],
    "Age":[20,21,22,19],
    "Marks":[70,85,75,90]
}
df=pd.DataFrame(data)
print("Original data:")
print(df)
#Filtering data
print("Alice's data:")
print(df[df["Name"]=='Alice'])
print("Students with marks less than 85:")
print(df[df["Marks"]<85])
print("Students with marks greater than 75 and age less than 20:")
print(df[(df["Age"]>20)&(df["Marks"]>75)])
```
